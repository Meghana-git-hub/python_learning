```python
def before_after(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper
@before_after
def say_hello():
    print("Hello")
say_hello()
```

```python
def start_decorator(func):
    def wrapper(name):
        print("Starting....")
        func(name)
        print("Done")
    return wrapper
@start_decorator
def greet(name):
    print("Hello",name)
greet("nani")
```

```python
def double_result(func):
    def wrapper():
        result=func()
        return result * 2
    return wrapper
@double_result
def get_value():
    return 25
print(get_value())
        
```

```python
def admin_only(func):
    def wrapper(user_role):
        if user_role !="admin":
            print("Access denied")
        else:
            func(user_role)
    return wrapper
@admin_only
def dashboard(role):
    print("Welcome admin")
dashboard("admin")
dashboard("Student")
        
```

```python
def to_upper(func):
    def wrapper():
        result=func()
        return result.upper()
    return wrapper
@to_upper
def get_msg():
    return "Hello World"
print(get_msg())
```

```python
def call_counter(func):
    count=0
    def wrapper():
        nonlocal count
        count+=1
        print(f"called {count} times")
        func()
    return wrapper
@call_counter
def greet():
    print("Hello")
greet()
greet()
```

```python
def add_prefix(func):
    def wrapper():
        result=func()
        return "ID:"+result
    return wrapper
@add_prefix
def get_name():
    return "Raju"
print(get_name())
```

```python
def message_wrapper(func):
    def wrapper():
        print("Initializing...")
        func()
        print("cleaning complete")
    return wrapper
@message_wrapper
def process():
    print("Function Logic Runs")
process()
```

```python
def prevent_negative(func):
    def wrapper(a,b):
        result = func(a,b)
        if result < 0:
            return 0
        return result
    return wrapper
@prevent_negative
def subtract(a,b):
    return a-b
print(subtract(5,10))
print(subtract(15,10))
```

```python
def validate_string(func):
    def wrapping(arg):
        if not isinstance(arg,str):
            print("Error:Invalid input type")
        else:
            func(arg)
    return wrapping
@validate_string
def print_text(text):
    print("you entered:",text)
print_text(123)
print_text("Hello")
```
