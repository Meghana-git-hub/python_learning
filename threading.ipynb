```python
from threading import Thread
import time
class myThread(Thread):
    def run(self):
        for i in range(10):
            time.sleep(1)
            print("starting  thread",i)
t1=myThread()
t2=myThread()
t1.start()
t2.start()
t1.join()
t2.join()
print("main thread")


            
        
    
```
