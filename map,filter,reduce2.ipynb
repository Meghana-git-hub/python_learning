```python
a=[1,2,3,4]
b=[10,20,30,40]
result=list(map(lambda x,y:x+y,a,b))
print(result)
```

```python
nums=[12,15,7,18,20,21,25]
result=list(filter(lambda x: (x % 3 == 0) != (x %5 ==0),nums))
print(result)
```

```python
from functools import reduce
nums=[1,2,3,4]
result=reduce(lambda x,y: x+y,nums,10)
print(result)
```

```python
nums=[[1,2],[3,4],[5,6]]
result=list(map(lambda x: x+[10],nums))
print("result:",result)
print("nums:",nums)
```
