```python
class Numbers:
    def __init__(self,n):
        self.n=n
        self.current=1
    def __iter__(self):
        return self
    def __next__(self):
        if self.current < self.n:
            value=self.current
            self.current+=1
            return value
        else:
            raise StopIteration
obj=Numbers(5)
for i in obj:
    print(i,end=" ")
```

```python
class Even:
    def __init__(self,n):
        self.n=n
        self.index=0
    def __iter__(self):
        return self
    def __next__(self):
        while self.index < len(self.n):
            value=self.n[self.index]
            self.index+=1
            if value%2==0:
                return value
        raise StopIteration
lst=[2,3,4,5,6,7]
for i in Even(lst):
    print(i,end=" ")
```

```python
class Rev:
    def __init__(self,text):
        self.text=text
        self.index=len(text)-1
    def __iter__(self):
        return self
    def __next__(self):
        if self.index >= 0:
            ch=self.text[self.index]
            self.index-=1
            return ch
        raise StopIteration
for ch in Rev("Python"):
    print(ch,end=" ")
            
    
```

```python
class IndexIterator:
    def __init__(self,data):
        self.data=data
        self.index=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.index < len(self.data):
            result=(self.index,self.data[self.index])
            self.index+=1
            return result
        raise StopIteration
lst=["python","is","a","programming","language"]
for i in IndexIterator(lst):
    print(i,end=" ")
```

```python
def digits(num):
    for d in str(num):
        yield int(d)
for i in digits(12345):
    print(i,end=" ")
```

```python
def cumulative_sum(lst):
    total=0
    for i in lst:
        total+=i
        yield total
for i in cumulative_sum([1,2,3,4,5]):
    print(i,end=" ")
```

```python
def vowels(text):
    for ch in text:
        if ch.lower() in "aeiou":
            yield ch
for i in vowels("Programming"):
    print(i,end=" ")
```

```python
class WordIterator:
    def __init__(self,sentence):
        self.words=sentence.split()
        self.index=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.index < len(self.words):
            word=self.words[self.index]
            self.index+=1
            return word
        raise StopIteration
for word in WordIterator("Python is easy"):
    print(word)
```

```python
class Evenchars:
    def __init__(self,text):
        self.text=text
        self.index=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.index < len(self.text):
            ch=self.text[self.index]
            self.index+=2
            return ch
        raise StopIteration
for ch in Evenchars("Python"):
    print(ch)
```

```python
def run_max(lst):
    max=lst[0]
    for i in lst:
        if i >max:
            max=i
        yield max
for i in run_max([9,4,6,7,10]):
    print(i)
```
