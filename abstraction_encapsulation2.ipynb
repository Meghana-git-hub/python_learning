```python
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimeter(self):
        pass
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return math.pi * self.r * self.r
    def perimeter(self):
        return 2 * math.pi * self.r
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l * self.b
    def perimeter(self):
        return 2 * (self.l + self.b)
class Triangle(Shape):
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def area(self):
        s = (self.a + self.b + self.c)/2
        return (s * (s-self.a) * (s-self.b) * (s-self.c)) ** 0.5
    def perimeter(self):
         return self.a + self.b + self.c
c=Circle(5)
print(c.area())
print(c.perimeter())
r=Rectangle(4,5)
print(r.area())
print(r.perimeter())
t=Triangle(2,3,4)
print(t.area())
print(t.perimeter())
    
    
```

```python
from abc import ABC, abstractmethod
class Paymentgateway(ABC):
    @abstractmethod
    def authenticate(self):
        pass
    def pay(self,amount):
        pass
    def refund(self,amount):
        pass
class UPIpayment(Paymentgateway):
    def authenticate(self):
        print("UPI Authenticated")
    def pay(self,amount):
        print("paid",amount,"using UPI")
    def refund(self,amount):
        print("Refunded",amount)
class Cardpayment(Paymentgateway):
    def authenticate(self):
        print("card authenticated")
    def pay(self,amount):
        print("paid",amount,"using UPI")
    def refund(self,amount):
        print("Refunded",amount)
class Netbankingpayment(Paymentgateway):
    def authenticate(self):
        print("card authenticated")
    def pay(self,amount):
        print("paid",amount,"using UPI")
    def refund(self,amount):
        print("Refunded",amount)
u=UPIpayment()
u.authenticate()
u.pay(1000)
u.refund(500)
cd=Cardpayment()
cd.authenticate()
cd.pay(100)
cd.refund(50)
nb=Netbankingpayment()
nb.authenticate()
nb.pay(2000)
nb.refund(1000)

    
```

```python
from abc import ABC, abstractmethod
class Vehiclecontrol(ABC):
    @abstractmethod
    def accelerate(self):
        pass
    def brake(self):
        pass
    def steer(self):
        pass
class Carcontrol(Vehiclecontrol):
    def accelerate(self):
        print("Car accelerating")
    def brake(self):
        print("Car braking")
    def steer(self):
        print("car steering")
class Bikecontrol(Vehiclecontrol):
    def accelerate(self):
        print("Bike accelerating")
    def brake(self):
        print("Bike braking")
    def steer(self):
        print("Bike steering")
class Truckcontrol(Vehiclecontrol):
    def accelerate(self):
        print("Truck accelerating")
    def brake(self):
        print("Truck braking")
    def steer(self):
        print("Truck steering")
car=Carcontrol()
car.accelerate()
car.brake()
car.steer()
bike=Bikecontrol()
bike.accelerate()
bike.brake()
bike.steer()
truck=Truckcontrol()
truck.accelerate()
truck.brake()
truck.steer()
```

```python
from abc import ABC, abstractmethod
class Databasedriver(ABC):
    @abstractmethod
    def connect(self):
        pass
    @abstractmethod
    def execute(self,query):
        pass
    @abstractmethod
    def close(self):
        pass
class Mysqldriver(Databasedriver):
    def connect(self):
        print("MYSQL connected")
    def execute(self,query):
        self.query=query
        print("Executing",query)
    def close(self):
        print("Closing MYSQL")
class Postgresdriver(Databasedriver):
    def connect(self):
        print("postgresql connected")
    def execute(self,query):
        self.query=query
        print("Executing",query)
    def close(self):
        print("Closing postgresql")
class sqllitedriver(Databasedriver):
    def connect(self):
        print("SQLLite connected")
    def execute(self,query):
        self.query=query
        print("Executing",query)
    def close(self):
        print("Closing SQLLite")
m=Mysqldriver()
m.connect()
m.execute("select * from student")
m.close()
p=Postgresdriver()
p.connect()
p.execute("select * from student")
p.close()
s=sqllitedriver()
s.connect()
s.execute("select * from student")
s.close()
```

```python
from abc import ABC, abstractmethod
class MLmodel(ABC):
    @abstractmethod
    def train(self,data):
        pass
    @abstractmethod
    def predict(self,x):
        pass
    @abstractmethod
    def evaluate(self,test_set):
        pass
class Linearregressionmodel(MLmodel):
    def train(self,data):
        print("Training linear regression model")
    def predict(self,x):
        print("predicting linear regression model")
    def evaluate(self,test_set):
        print("Evaluating linear regression model")
class Decisiontreemodel(MLmodel):
    def train(self,data):
        print("Training Decision tree model")
    def predict(self,x):
        print("Predicting decision tree model")
    def evaluate(self, test_set):
        print("Evaluating decision tree model")
l=Linearregressionmodel()
l.train("Hello")
l.predict(5)
l.evaluate("Test_data")
d=Decisiontreemodel()
d.train("Hi")
d.predict(4)
d.evaluate("test_data")
```

```python
from abc import ABC, abstractmethod
class Notifier(ABC):
    @abstractmethod
    def send(self,message):
        pass
class Emailsender(Notifier):
    def send(self,message):
        print("Email:",message)
class SMSsender(Notifier):
    def send(self,message):
        print("SMS:",message)
class Pushsender(Notifier):
    def send(self,message):
        print("Push:",message)
e=Emailsender()
e.send("Welcome")
s=SMSsender()
s.send("Hey Guys")
p=Pushsender()
p.send("Good morning")
    
```
