```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def is_passed(self):
        return self.marks > 40


s1 = Student("A", 60)
s2 = Student("B", 30)

print(s1.name, s1.is_passed())
print(s2.name, s2.is_passed())
        
```

```python
class Employee:
    company_name="TechCorp"
    def __init__(self,name):
        self.name=name
    @classmethod
    def change_company(cls,new_name):
        cls.company_name=new_name
e1=Employee("Carrie")
e2=Employee("David")
print(Employee.company_name)
Employee.change_company("CorpTech")
print(Employee.company_name)
print(e1.name,e1.company_name)
print(e2.name,e2.company_name)
```

```python
class MathOps:
    @staticmethod
    def is_even(num):
        return num%2==0
print(MathOps.is_even(4))
m=MathOps()
print(m.is_even(5))
```

```python
class car:
    wheels=4
    def __init__(self,mileage):
        self.mileage=mileage
    @classmethod
    def change_wheels(cls,new_wheels):
        cls.wheels=new_wheels
    def display_specs(self):
        print(self.mileage,self.wheels)
c=car(20)
c.display_specs()
car.change_wheels(10)
c.display_specs()
```

```python
class Temperature:
    def __init__(self,celsius):
        self.celsius=celsius
    @staticmethod
    def to_fahrenheit(c):
        return (c * 9/5)+32
    def show(self):
        print(self.celsius,Temperature.to_fahrenheit(self.celsius))
t=Temperature(25)
t.show()
        
```

```python
class Employee:
    bonus_rate=0.1
    def __init__(self,name,base_salary):
        self.name=name
        self.base_salary=base_salary
    def final_salary(self):
        return self.base_salary + (self.base_salary * self.bonus_rate)
    @classmethod
    def update_bonus(cls,new_rate):
        cls.bonus_rate=new_rate
    @staticmethod
    def is_valid_salary(sal):
        return sal > 0
e1=Employee("Emma",10000)
e2=Employee("Farina",20000)
print(e1.final_salary(),e2.final_salary())
Employee.update_bonus(0.2)
print(e1.final_salary(),e2.final_salary())
        
        
        
```

```python
class course:
    total_students=0
    def __init__(self,std_name):
        self.std_name=std_name
    def enroll(self):
        course.total_students+=1
    @classmethod
    def show_total(cls):
        print("total_students:",cls.total_students)
    @staticmethod
    def is_eligible(age):
        return age > 18
c1=course("Anu")
c2=course("Hanu")
c1.enroll()
c2.enroll()
course.show_total()
```

```python
class Bankaccount:
    bank_name="SBI"
    def __init__(self,holder,balance):
        self.holder=holder
        self.balance=balance
    def deposit(self,amount):
        if Bankaccount.validate_amount(amount):
            self.balance+=amount
    @classmethod
    def change_bank_name(cls,new_name):
        cls.bank_name=new_name
    @staticmethod
    def validate_amount(amount):
        return amount > 0
acc=Bankaccount("Harry",2000)
acc.deposit(1000)
print(acc.holder,acc.balance,Bankaccount.bank_name)
Bankaccount.change_bank_name("HDFC")
print(Bankaccount.bank_name)
```

```python
class Student:
    passing_marks=40
    def __init__ (self,name,marks):
        self.name=name
        self.marks=marks
    def result(self):
        if self.marks >= Student.passing_marks:
            print(self.name,"Pass")
        else:
            print(self.marks,"Fail")
    @classmethod
    def update_passing_marks(cls,new_marks):
        cls.passing_marks=new_marks
    @staticmethod
    def grade_category(marks):
        if marks >= 75:
            return "A"
        elif marks >=50:
            return "B"
        else:
            return "c"
s1=Student("Ayan",60)
s2=Student("poorna",90)
s3=Student("Hari",30)
s1.result()
s2.result()
s3.result()
Student.update_passing_marks(50)
print(s1.name,Student.grade_category(s1.marks))
print(s2.name,Student.grade_category(s2.marks))
print(s3.name,Student.grade_category(s3.marks))
```
