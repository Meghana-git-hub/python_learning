```python
from abc import ABC, abstractmethod
class notifier(ABC):
    @abstractmethod
    def send(self,message):
        pass
class Emailnotifier(notifier):
    def __init__(self):
        self.__api_key="EMAIL123"
    def send(self,message):
        print(f"Email sent:{message}")
class Smsnotifier(notifier):
    def __init__(self):
        self.__api_key="SMS123"
    def send(self,message):
        print(f"SMS sent:{message}")
class Pushnotifier(notifier):
    def __init__(self):
        self.__api_key="PUSH123"
    def send(self,message):
        print(f"Push notification sent:{message}")
def notify_users(notifiers,message):
    for notifier in notifiers:
        notifier.send(message)
notifications=[Emailnotifier(),Smsnotifier(),Pushnotifier()]
notify_users(notifications,"Welcome to our platform!")
        
```

```python
from abc import ABC, abstractmethod
class paymentprocessor(ABC):
    def __validate(self,amount):
        if amount <= 0:
            raise ValueError("Invalid amount")
        return True
    def validate_payment(self,amount):
        return self.__validate(amount)
    @abstractmethod
    def process_payment(self,amount):
        pass
class Creditcardpayment(paymentprocessor):
    def process_payment(self,amount):
        self.validate_payment(amount)
        print(f"Credit card payment of {amount} processed")
class UPIpayment(paymentprocessor):
    def process_payment(self,amount):
        self.validate_payment(amount)
        print(f"UPI payment of {amount} processed")
class Walletpayment(paymentprocessor):
    def process_payment(self,amount):
        self.validate_payment(amount)
        print(f"Wallet payment of {amount} processed")
def make_payment(payment,amount):
    payment.process_payment(amount)
make_payment(Creditcardpayment(),5000)
make_payment(UPIpayment(),15000)
make_payment(Walletpayment(),25000)
    
```

```python
class Filewriter:
    def write(self,message):
        print(f"Writing to file:{message}")
class Databasewriter:
    def write(self,message):
        print(f"Writing to Database:{message}")
class Monitoringwriter:
    def write(self,message):
        print(f"sending to monitoring system:{message}")
class Logger:
    def __init__(self,writer):
        self.writer = writer
    def set_writer(self,writer):
        self.writer = writer
    def log(self,message):
        self.writer.write(message)
logger=Logger(Filewriter())
logger.log("Application Started")
logger.set_writer(Databasewriter())
logger.log("user logged in")
logger.set_writer(Monitoringwriter())
logger.log("Server Error detected")
```

```python
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length=length
        self.width=width
    def area(self):
        return self.length * self.width
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return math.pi * self.radius * self.radius
class Triangle(Shape):
    def __init__(self, base, height):
        self.base=base
        self.height=height
    def area(self):
        return 0.5 * self.base * self.height
shapes=[Rectangle(10,5),Triangle(8,4),Circle(5)]
for shape in shapes:
    print("Area=",shape.area())
    
```

```python
class Baseprice:
    def calculate(self,price):
        print("Base price:",price)
        return price
class Discount:
    def calculate(self,price):
        price = price - (price * 10/100)
        print("After Discount:",price)
        return price
class Tax:
    def calculate(self,price):
        price = price + (price * 18/100)
        print("After tax:",price)
        return price
class Order(Discount,Tax,Baseprice):
    def final_price(self,price):
        return self.calculate(price)
order=Order()
print("final price:",order.final_price(1000))
print("MRO:",Order.__mro__) 
    
```

```python
class Bankaccount:
    def __init__(self,balance):
        self.__balance=balance
    @property
    def balance(self):
        return self.__balance
    @balance.setter
    def balance(self,amount):
        if amount >= 0:
            self.__balance=amount
        else:
            print("Invalid Balance! Balance cannot be negative")
acc=Bankaccount(10000)
print("current balance:",acc.balance)
acc.balance=15000
print("Updated balance:",acc.balance)
acc.balance = -500
print("Final balance:",acc.balance)
```

```python
from abc import ABC, abstractmethod
class vehicle(ABC):
    @abstractmethod
    def move(Self):
        pass
class petrolcar(vehicle):
    def move(self):
        print("Petrol car is running using an engine")
class Electriccar(vehicle):
    def move(self):
        print("Electric car is running using a battery")
class Bicycle(vehicle):
    def move(self):
        print("Bicycle is moving using pedals")
vehicles=[petrolcar(),Electriccar(),Bicycle()]
for vehicle in vehicles:
    vehicle.move()
```

```python
from abc import ABC, abstractmethod
class Authenticate(ABC):
    def __init__(self,username,password):
        self.username=username
        self.__password=password
    def validate(self):
        return len(self.__password) >= 6
    @abstractmethod
    def login(self):
        pass
class Passwordauth(Authenticate):
    def login(self):
        if self.validate():
            print(f"{self.username}logged in using password authentication")
        else:
            print("Invalid password")
class Otpauth(Authenticate):
    def login(self):
        if self.validate():
            print(f"{self.username} logged in using otp authentication")
        else:
            print("Invalid OTP")
class Biometricauth(Authenticate):
    def login(self):
        if self.validate():
            print(f"{self.username} logged in using biometric authentication")
        else:
            print("Authentication failed")
users=[Passwordauth("Emma","abc123"),Otpauth("Rahul","654321"),Biometricauth("Anjali","bio123")]
for user in users:
    user.login()
    
```
