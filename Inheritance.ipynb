```python
class Animal:
    def sound(self):
        print("Animal makes a sound")
class Dog(Animal):
    def sound(self):
        print("Dog barks")
d=Dog()
d.sound()
        
```

```python
class A:
    def show(self):
        print("class-A:Hey guys")
class B(A):
    def show(self):
        super().show()
        print("class-B:Good morning to all")
obj=B()
obj.show()
        
```

```python
class Vehicle:
    pass
class Car(Vehicle):
    def wheels(self):
        print("Car has 4 wheels")
class Bike(Vehicle):
    def wheels(self):
        print("Bike has 2 wheels")
obj=Car()
obj.wheels()
obj2=Bike()
obj2.wheels()
```

```python
class Employee:
    def salary(self):
        print("Employee salary starts from 20k")
class Manager(Employee):
    def salary(self):
        super().salary()
        print("Manager salary with incentive 50k")
e=Employee()
e.salary()
m=Manager()
m.salary()
```

```python
class University:
    name="ABC University"
    @classmethod
    def show_university(cls):
        print(cls.name)
class College(University):
    pass
print(College.name)
College.show_university()
```

```python
class Mathops:
    @staticmethod
    def add(a,b):
        return a+b
class Advancedops(Mathops):
    pass
print(Advancedops.add(5,3))    
```

```python
from abc import ABC, abstractmethod
class Shape:
    @abstractmethod
    def area(self):
        pass
class rectangle(Shape):
    def area(self):
        return 10*5
r=rectangle()
print(r.area())
```
