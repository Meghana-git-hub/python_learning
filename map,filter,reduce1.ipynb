```python
l=[[1,2],[3,4],[5,6]]
k=list(map(lambda x:x+[5],l))
print(k)
```

```python
d={"apple":100,"banana":40,"cherry":150}
k=list(filter(lambda x:x[1]>50,d.items()))
print(k)
```

```python
from functools import reduce
l=[20,30,6,9,40]
k=reduce(lambda x,y: x if x>y else y,l)
print(k)
```

```python
s="hello, Who are you?"
x="AEIOU aeiou"
k=list(filter(lambda ch: ch not in x,s))
print(" ".join(k))
```

```python
from functools import reduce
l=['p','y','t','h','o','n']
k=reduce(lambda x,y:x+y,l)
print(k)
```

```python
l=[13,24,56,76.90]
k=list(map(lambda x:id(x),l))
print(k)
```

```python
from functools import reduce
l=[5,10,15,20,25,30]
k=reduce(lambda x,y:x+y,filter(lambda x:x%5==0,map(lambda x:x**x,l)))
print(k)
```
