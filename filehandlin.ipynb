```python
from contextlib import contextmanager
@contextmanager
def open_file(filename,mode):
    file=open(filename,mode)
    try:
        yield file
    finally:
        file.close()
        print("file closed")
with open_file("sample.txt","r") as f:
    print(f.read())
```

```python
with open("sample.txt","r") as file:
    while True:
        chunk=file.read(5)
        if not chunk:
            break
        print("data:",chunk)
        print("cursor position:",file.tell())
        print("\n moving cursor to position 10")
        file.seek(10)
        print("reading from position 10:")
        print(file.read())
```
