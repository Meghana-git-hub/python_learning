```python
def sum_even(numbers):
    total=0
    for num in numbers:
        if num%2==0:
            total+=num
    return total
num=[1,12,3,14,5]
print(sum_even(num))
```

```python
def count_vowels(text):
    vowels="aeiouAEIOU"
    count=0
    for ch in text:
        if ch in vowels:
            count+=1
    return count
print(count_vowels("Hello World"))
```

```python
def student_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
student_details(name="Anu",age=22,course="python")
student_details(name="Abhi",age=22,course="java")
```

```python
def outer():
    x=10
    def inner():
        nonlocal x
        x+=5
        print("Inside:",x)
    inner()
    print("Outside:",x)
outer()
```

```python
celsius=[0,20,30,40]
fahrenheit=list(map(lambda c:(c*9/5)+32,celsius))
print(fahrenheit)
```

```python
names=["Anu","Bob","Carrie","David"]
first_letter=list(map(lambda name:name[0],names))
print(first_letter)
```

```python
words=["Apple","carrot","orange","Guava"]
result=list(filter(lambda word:word[0].lower() in "aeiou",words))
print(result)
```

```python
from functools import reduce
nums=[1,2,3,4,5,10,15]
even=list(filter(lambda x: x%2==0,nums))
product=reduce(lambda x,y: x*y,even)
print(product)
```

```python
n=5
factorial=reduce(lambda x,y:x*y,range(1,n+1))
print(factorial)
```

```python
nums=[1,2,3,4,5,6,7]
odd=filter(lambda x: x%2!=0,nums)
sum=reduce(lambda x,y:x+y,odd)
print(sum)
```
