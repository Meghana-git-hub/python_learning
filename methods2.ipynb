```python
class Student:
    total_students=0
    pass_marks=40
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
        Student.total_students+=1
    def is_pass(self):
        return self.marks >= Student.pass_marks
    def curve_marks(self,percent):
        self.marks+=self.marks*percent/100
    @staticmethod
    def update_marks(marks):
        if marks > 90:
            return "A"
        elif marks > 70:
            return "B"
        elif marks > 40:
            return "c"
        else:
            return "D"
s1=Student("Anu",15)
s2=Student("Bob",70)
s1.curve_marks(20)
s2.curve_marks(10)
for s in [s1,s2]:
    print(s.name,s.marks,Student.update_marks(s.marks),"pass" if s.is_pass() else "Fail")

        
    
```

```python
class product:
    base_tax=0.1
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def final_price(self):
        return self.price+(self.price*product.base_tax)
    @classmethod
    def update_tax(cls,new_tax):
        cls.base_tax=new_tax
    @staticmethod
    def is_valid(price):
        return price >= 0 and price < 50000
p1=product("Mobile",10000)
p2=product("Laptop",50000)
product.update_tax(0.2)
for p in [p1,p2]:
    print(p.name,p.final_price(),product.is_valid(p.price))
```

```python
class Employee:
    min_exp=5
    valid_dept=["HR","Tech","Admin"]
    def __init__(self,ename,exp,dept):
        self.ename=ename
        self.exp=exp
        self.dept=dept
    def check_eligibility(self):
        return self.exp > Employee.min_exp
    @classmethod
    def update_exp(cls,new_exp):
        cls.min_exp=new_exp
    @staticmethod
    def is_valid_dept(dept):
        return dept in Employee.valid_dept
e1=Employee("Emma",7,"HR")
e2=Employee("Farin",4,"Tech")
Employee.update_exp(6)
for e in [e1,e2]:
    print(e.ename,e.check_eligibility(),Employee.is_valid_dept(e.dept))
```

```python
class loan:
    intrest_rate=0.1
    def __init__(self,bname,principal):
        self.bname=bname
        self.principal=principal
    def total_amount(self):
        return self.principal+(self.principal*loan.intrest_rate)
    @classmethod
    def update_intrest(cls,new_rate):
        cls.intrest_rate=new_rate
    @staticmethod
    def loan_eligible(salary):
        return salary > 20000
l1=loan("Gani",10000)
l2=loan("Harry",12000)
loan.update_intrest(0.2)
for l in [l1,l2]:
    print(l.bname,l.principal,loan.loan_eligible(25000),l.total_amount())    
```

```python
class course:
    total_courses=0
    min_duration=1
    def __init__(self,title,duration):
        self.title=title
        self.duration=duration
        self.students=0
        course.total_courses+=1
    def enroll(self):
        self.students+=1
    @classmethod
    def update_duration(cls,new_duration):
        cls.min_duration=new_duration
    @staticmethod
    def check_duration(duration):
        return duration > 0 and duration < 20
c1=course("python",6)
c2=course("Java",20)
c1.enroll()
c2.enroll()
course.update_duration(15)
for c in [c1,c2]:
    print(c.title,c.students,course.check_duration(c.duration))
        
```

```python
class vehicle:
    service_rate=2
    def __init__(self,model,k_run):
        self.model=model
        self.k_run=k_run
    def cal_service(self):
        return self.k_run*vehicle.service_rate
    @classmethod
    def update_service(cls,new_service):
        cls.service_rate=new_service
    @staticmethod
    def check_eligible(year):
        return year >2008
v1=vehicle("car",1000)
v2=vehicle("bike",200)
vehicle.update_service(4)
for v in [v1,v2]:
    print(v.model,v.cal_service(),vehicle.check_eligible(2009))
        
```

```python
class inventory:
    total_items=0
    min_threshold=5
    def __init__(self):
        self.stock={}
    def add_item(self,item,qty):
        self.stock[item]=self.stock.get(item,0)+qty
        inventory.total_items+=1
    def remove_item(self,item,qty):
        if item in self.stock:
            self.stock[item]-=qty
            inventory.total_items-=1
    @classmethod
    def update_threshold(cls,val):
        cls.min_threshold=val
    @staticmethod
    def below_threshold(qty):
        return qty < inventory.min_threshold
i1=inventory()
i1.add_item("pen",10)
i2=inventory()
i2.add_item("Book",50)
for i in [i1,i2]:
    print(i.stock,inventory.below_threshold(3))
```

```python
class hotelroom:
    base_price=5000
    def __init__(self,rno,nyts,name):
        self.rno=rno
        self.nyts=nyts
        self.name=name
    def total_bill(self):
        return self.nyts*hotelroom.base_price
    @classmethod
    def update_price(cls,new_price):
        cls.base_price=new_price
    @staticmethod
    def check_utility(n):
        return isinstance(n,int) and n>0
h1=hotelroom(101,2,"Jimmy")
h2=hotelroom(105,3,"Kevin")
hotelroom.update_price(10000)
for h in [h1,h2]:
    print(h.rno,h.total_bill(),hotelroom.check_utility(3))
    
               
        
```

```python
class librarymember:
    total_members=0
    books_limit=3
    def __init__(self,name):
        self.name=name
        self.books=0
        librarymember.total_members+=1
    def borrow(self):
        if self.books < librarymember.books_limit:
            self.books+=1
    @classmethod
    def update_limit(cls,limit):
        cls.books_limit=limit
    @staticmethod
    def valid_title(title):
        return isinstance(title,str) and len(title)>0
m1=librarymember("Leela")
m2=librarymember("manu")
m1.borrow()
m2.borrow()
for m in [m1,m2]:
    print(m.name,m.books,librarymember.valid_title("Python"))
```

```python
class member:
    bmi_limit=25
    def __init__(self,name,height,weight):
        self.name=name
        self.height=height
        self.weight=weight
    def bmi(self):
        return self.weight/(self.height**2)
    def is_fit(self):
        return self.bmi()<=member.bmi_limit
    @classmethod
    def update_limit(cls,limit):
        cls.bmi_limit=limit
    @staticmethod
    def valid_input(h,w):
        return  h>0 and w>0
m1=member("Nani",1.6,55)
m2=member("Iyan",5.5,60)
for m in [m1,m2]:
    print(m.bmi(),m.is_fit(),member.valid_input(1.6,60))
```
